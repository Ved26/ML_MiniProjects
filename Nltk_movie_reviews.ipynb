```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
```

```python
from nltk.corpus import stopwords
import string
stops = set(stopwords.words('english'))
punctuations = list(string.punctuation)
stops.update(punctuations)
stops
```

```python
from nltk import pos_tag
```

```python
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
def clean_words(words):
    output_words = []
    for w in words:
        if w.lower() not in stops:
            pos = pos_tag([w])
            clean_word = lemmatizer.lemmatize(w,pos=get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words 
```

```python
documents = [(clean_words(document),category)for document,category in documents]
```

```python
documents[0]
```

```python
training_documents = documents[0:1500]
testing_documents = documents[1500:]
```

```python
all_words = []
for doc in training_documents:
    all_words += doc[0]
```

```python
import nltk
```

```python
freq = nltk.FreqDist(all_words)
common = freq.most_common(3000)
features = [i[0] for i in common]
```

```python
features
```

```python
def get_feature_dict(words):
    current_features= {}
    words_set = set(words)
    for w in features:
        current_features[w] = w in words_set
    return current_features
```

```python
training_data = [(get_feature_dict(doc),category) for doc,category in training_documents]
testing_data = [(get_feature_dict(doc),category) for doc,category in training_documents]
```

```python
training_data[0]
```

```python
classifier = NaiveBayesClassifier.train(training_data)
```

```python
nltk.classify.accuracy(classifier,testing_data)
```

```python
svc = SVC()
svc.fit(x_train_features,y_train)
```

```python
svc.score(x_test_features,y_test)
```

```python
y_pred = svc.predict(x_test_features)
```

```python
from sklearn.metrics import confusion_matrix
```
